```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
well_0 = pd.read_csv("well_0.csv")
```

.describe() function to view statistics

```python
well_0.describe()
```

```python
fig, ax = plt.subplots(figsize=(20,20))

#Set up the plot axes
ax1 = plt.subplot2grid((1,9), (0,0), rowspan=1, colspan = 1) # Bit Size
ax3 = plt.subplot2grid((1,9), (0,1), rowspan=1, colspan = 1) # Gamma-ray
ax4 = plt.subplot2grid((1,9), (0,2), rowspan=1, colspan = 1) # Density
ax6 = plt.subplot2grid((1,9), (0,3), rowspan=1, colspan = 1) # Density correction
ax8 = plt.subplot2grid((1,9), (0,4), rowspan=1, colspan = 1) # PEF
ax10 = plt.subplot2grid((1,9), (0,5), rowspan=1, colspan = 1) # Deep resistivity
ax12 = plt.subplot2grid((1,9), (0,6), rowspan=1, colspan = 1) # Final porosity
ax13 = plt.subplot2grid((1,9), (0,7), rowspan=1, colspan = 1) # Water saturation
ax14 = plt.subplot2grid((1,9), (0,8), rowspan=1, colspan = 1) # Shale volume
ax2 = ax1.twiny() #Twins the y-axis for the bit size track with the caliper track, it creates a twin Axes sharing the yaxis.
ax5 = ax4.twiny() #Twins the y-axis for the density track with the neutron track, it creates a twin Axes sharing the yaxis.
ax7 = ax6.twiny() #Twins the y-axis for the ROP with the density correcction track, it creates a twin Axes sharing the yaxis.
ax9 = ax8.twiny() #Twins the y-axis for the sonic track with the PEF track, it creates a twin Axes sharing the yaxis.
ax11 = ax10.twiny() #Twins the y-axis for the deep resistivity track with the medium resistivity track, it creates a twin Axes sharing the yaxis.

plt.suptitle("Well 0", fontsize=24)

# As our curve scales will be detached from the top of the track,
# this code adds the top border back in without dealing with splines
ax15 = ax1.twiny()
ax15.xaxis.set_visible(False)
ax16 = ax3.twiny()
ax16.xaxis.set_visible(False)
ax17 = ax4.twiny()
ax17.xaxis.set_visible(False)
ax18 = ax6.twiny()
ax18.xaxis.set_visible(False)
ax19 = ax8.twiny()
ax19.xaxis.set_visible(False)
ax20 = ax10.twiny()
ax20.xaxis.set_visible(False)
ax21 = ax12.twiny()
ax21.xaxis.set_visible(False)
ax22 = ax13.twiny()
ax22.xaxis.set_visible(False)
ax23 = ax14.twiny()
ax23.xaxis.set_visible(False)

# Bit Size track
ax1.plot("BS", "DEPTH", data = well_0, color = "black")
ax1.set_xlabel("Bit Size (inches)")
ax1.xaxis.label.set_color("black")
ax1.set_xlim(6, 22)
ax1.set_ylabel("Depth (m)")
ax1.tick_params(axis='x', colors="black")
ax1.spines["top"].set_edgecolor("black")
ax1.title.set_color('black')
ax1.set_xticks([6, 10, 14, 18, 22])

# Caliper placed ontop of Bit Size track
ax2.plot("CALI", "DEPTH", data = well_0, color = "blue")
ax2.set_xlabel("Caliper (inches)")
ax2.xaxis.label.set_color("blue")
ax2.set_xlim(6, 22)
ax2.set_ylim(15250, 11000)
ax2.tick_params(axis='x', colors="blue")
ax2.spines["top"].set_position(("axes", 1.08))
ax2.spines["top"].set_visible(True)
ax2.spines["top"].set_edgecolor("blue")
ax2.set_xticks([6, 10, 14, 18, 22])

# Gamma Ray track
ax3.plot("GR", "DEPTH", data = well_0, color = "green")
ax3.set_xlabel("Gamma-ray (API)")
ax3.xaxis.label.set_color("green")
ax3.set_xlim(0, 350)
ax3.set_ylabel("Depth (m)")
ax3.tick_params(axis='x', colors="green")
ax3.spines["top"].set_edgecolor("green")
ax3.title.set_color('green')
ax3.set_xticks([0, 100, 200, 300])

# Density track
ax4.plot("DEN", "DEPTH", data = well_0, color = "red")
ax4.set_xlabel("Bulk density (g/cc)")
ax4.set_xlim(1.9, 3)
ax4.xaxis.label.set_color("red")
ax4.tick_params(axis='x', colors="red")
ax4.spines["top"].set_edgecolor("red")
#ax4.set_xticks([1.9, 2.175, 2.45, 2.725, 3])

# Neutron track placed ontop of density track
ax5.plot("NEU", "DEPTH", data = well_0, color = "blue")
ax5.set_xlabel("Neutron (v/v)")
ax5.xaxis.label.set_color("blue")
ax5.set_xlim(0.45, -0.15)
ax5.set_ylim(15250, 11000)
ax5.tick_params(axis='x', colors="blue")
ax5.spines["top"].set_position(("axes", 1.08))
ax5.spines["top"].set_visible(True)
ax5.spines["top"].set_edgecolor("blue")
ax5.set_xticks([0.45, 0.3, 0.15,0, -0.15])

x4=well_0['DEN']
x5=well_0['NEU']

x = np.array(ax4.get_xlim())
z = np.array(ax5.get_xlim())

nz=((x5-np.max(z))/(np.min(z)-np.max(z)))*(np.max(x)-np.min(x))+np.min(x)

ax4.fill_betweenx(well_0['DEPTH'], x4, nz, where=x4>=nz, interpolate=True, color='green')
ax4.fill_betweenx(well_0['DEPTH'], x4, nz, where=x4<=nz, interpolate=True, color='yellow')

# Density correction track
ax6.plot("DENC", "DEPTH", data = well_0, color = "black")
ax6.set_xlabel("Density correction (g/cc)")
ax6.set_xlim(-0.75, 0.25)
ax6.xaxis.label.set_color("black")
ax6.tick_params(axis='x', colors="black")
ax6.spines["top"].set_edgecolor("black")
ax6.set_xticks([-0.75, -0.5, -0.25, 0, 0.25])

# ROP track placed ontop of density correction track
ax7.plot("ROP", "DEPTH", data = well_0, color = "red")
ax7.set_xlabel("ROP (m/h)")
ax7.xaxis.label.set_color("red")
ax7.set_xlim(250, 0)
ax7.set_ylim(15250, 11000)
ax7.tick_params(axis='x', colors="red")
ax7.spines["top"].set_position(("axes", 1.08))
ax7.spines["top"].set_visible(True)
ax7.spines["top"].set_edgecolor("red")
ax7.set_xticks([250, 200, 150, 100, 50, 0])

# PEF track
ax8.plot("PEF", "DEPTH", data = well_0, color = "purple")
ax8.set_xlabel("PEF (b/e)")
ax8.set_xlim(2, 12)
ax8.xaxis.label.set_color("purple")
ax8.tick_params(axis='x', colors="purple")
ax8.spines["top"].set_edgecolor("purple")
ax8.set_xticks([2, 4, 6, 8, 10, 12])

# Sonic track placed ontop of PEF track
ax9.plot("DTC", "DEPTH", data = well_0, color = "blue")
ax9.set_xlabel("Sonic (usec/ft)")
ax9.xaxis.label.set_color("blue")
ax9.set_xlim(200, 0)
ax9.set_ylim(15250, 11000)
ax9.tick_params(axis='x', colors="blue")
ax9.spines["top"].set_position(("axes", 1.08))
ax9.spines["top"].set_visible(True)
ax9.spines["top"].set_edgecolor("blue")
ax9.set_xticks([200, 150, 100, 50, 0])

# Deep resistivity track
ax10.plot("RDEP", "DEPTH", data = well_0, color = "blue")
ax10.set_xlabel("Rdeep (ohm.m)")
ax10.xaxis.label.set_color("blue")
ax10.set_xlim(0.2, 200)
ax10.tick_params(axis='x', colors="blue")
ax10.spines["top"].set_edgecolor("blue")
ax10.title.set_color('blue')
ax10.semilogx()

# Medium resistivity track placed ontop of Deep resistivity track
ax11.plot("RMED", "DEPTH", data = well_0, color = "magenta")
ax11.set_xlabel("Rmed (ohm.m)")
ax11.xaxis.label.set_color("magenta")
ax11.set_xlim(0.2, 200)
ax11.set_ylim(15250, 11000)
ax11.tick_params(axis='x', colors="magenta")
ax11.spines["top"].set_position(("axes", 1.08))
ax11.spines["top"].set_visible(True)
ax11.spines["top"].set_edgecolor("magenta")
ax11.semilogx()

# Final porosity track
ax12.plot("PHIF", "DEPTH", data = well_0, color = "black")
ax12.set_xlabel("Final porosity (v/v)")
ax12.xaxis.label.set_color("black")
ax12.set_xlim(0,0.5)
ax12.tick_params(axis='x', colors="black")
ax12.spines["top"].set_edgecolor("black")
ax12.title.set_color('black')
ax12.set_xticks([0, 0.25, 0.5])

# Water saturation track
ax13.plot("SW", "DEPTH", data = well_0, color = "blue")
ax13.set_xlabel("Water saturation (v/v)")
ax13.xaxis.label.set_color("blue")
ax13.set_xlim(0,1)
ax13.tick_params(axis='x', colors="blue")
ax13.spines["top"].set_edgecolor("blue")
ax13.title.set_color('blue')
ax13.set_xticks([0, 0.25, 0.5, 0.75, 1])

# Shale volume track
ax14.plot("VSH", "DEPTH", data = well_0, color = "green")
ax14.set_xlabel("Shale volume (v/v)")
ax14.xaxis.label.set_color("green")
ax14.set_xlim(0,1)
ax14.tick_params(axis='x', colors="green")
ax14.spines["top"].set_edgecolor("green")
ax14.title.set_color('green')
ax14.set_xticks([0, 0.25, 0.5, 0.75, 1])

# Common functions for setting up the plot can be extracted into
# a for loop. This saves repeating code.
for ax in [ax1, ax3, ax4, ax6, ax8, ax10, ax12, ax13, ax14]:
    ax.set_ylim(15250, 11500)
    ax.grid(which='major', color='lightgrey', linestyle='-')
    ax.xaxis.set_ticks_position("top")
    ax.xaxis.set_label_position("top")
    ax.spines["top"].set_position(("axes", 1.02))

#plt.savefig('well_0.png')
```
