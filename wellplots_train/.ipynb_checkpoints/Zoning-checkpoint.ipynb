```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Definition of Well_0 depth of interest

```python
from pathlib import Path
data_folder = Path("/Users/djnav/OneDrive/Documentos/GitHub/SPWLA_2021_Geolatinas/dataset/separated_wells/train")
well_0_file = data_folder / "well_0.csv"
well_0 = pd.read_csv(well_0_file, index_col=0)
well_0_PHIF = well_0.loc[well_0.PHIF.notnull()]
well_0_SW = well_0.loc[well_0.SW.notnull()]
well_0_VSH = well_0.loc[well_0.VSH.notnull()]
depth_comparison_0 = pd.DataFrame({'PHIF':[well_0_PHIF.DEPTH.max(),well_0_PHIF.DEPTH.min()],'SW':[well_0_SW.DEPTH.max(),well_0_SW.DEPTH.min()],'VSH':[well_0_VSH.DEPTH.max(),well_0_VSH.DEPTH.min()]}, index=['BOTTOM', 'TOP'])
depth_comparison_0
```

```python
well_0.describe()
```

```python
well_0.shape
```

```python
#Searching for the index location of the top depth of interest
well_0.DEPTH.iloc[22748]
```

```python
#Final dataframe to use for training
final_well_0 = well_0[22748:29754]
final_well_0
```

```python
#save as CSV file
final_well_0.to_csv('well0_depth.csv')
```
