```python
GR_min = well_1_VSH.GR.min()
GR_min
```

```python
GR_max = well_1_VSH.GR.max()
GR_max
```

```python
#Define Vshal lineal function
def shale_volume(gamma_ray, gamma_ray_max, gamma_ray_min):
    vshale = (gamma_ray - gamma_ray_min) / (gamma_ray_max - gamma_ray_min)
    return round(vshale, 4)
```

```python
#Testing fomula
shale_volume(120, GR_max, GR_min)
```

```python
#Calculate Shale Volume
VSH_Lineal1 = shale_volume(well_1_VSH['GR'], 250, 20)
VSH_Lineal1.describe()
```

```python
well_1_final= well_1_VSH.assign(VSH_Lineal = VSH_Lineal1)
```

```python
well_1_final.describe()
```
